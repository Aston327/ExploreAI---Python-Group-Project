```python
# Number of unique values for each PLU code
unique_counts = [avocado_data['plu4046'].nunique(), avocado_data['plu4225'].nunique(), avocado_data['plu4770'].nunique()]
plu_codes = ['PLU 4046', 'PLU 4225', 'PLU 4770']

# Bar plot
plt.figure(figsize=(6, 6))
barplot = sns.barplot(x=plu_codes, y=unique_counts, hue=plu_codes, palette='viridis', legend=False)
plt.title('Number of Unique Values for PLU Codes')
plt.xlabel('PLU Codes')
plt.ylabel('Number of Unique Values')

# Add data values on top of the bars
for i, count in enumerate(unique_counts):
    barplot.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```

```python
#Total Sales Volume by PLU Code:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your avocado data
avocado_data = pd.read_csv('Avocado_HassAvocadoBoard_20152023v1.0.1.csv')

# Calculate total sales volume for each PLU code
plu_codes = ['PLU 4046', 'PLU 4225', 'PLU 4770']
plu_columns = ['plu4046', 'plu4225', 'plu4770']
total_sales = {plu: avocado_data[column].sum() for plu, column in zip(plu_codes, plu_columns)}

# Create a bar plot for total sales volume
plt.figure(figsize=(6, 6))
barplot = sns.barplot(x=list(total_sales.keys()), y=list(total_sales.values()), hue=list(total_sales.keys()), palette='viridis', legend=False)
plt.title('Total Sales Volume for PLU Codes')
plt.xlabel('PLU Codes')
plt.ylabel('Total Sales Volume')

# Add data values on top of the bars
for i, (plu, sales) in enumerate(total_sales.items()):barplot.text(i, sales + 0.1, f'{sales}', ha='center', va='bottom')
plt.show()
```

```python
#Total Average Price for PLU Code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your avocado data
avocado_data = pd.read_csv('Avocado_HassAvocadoBoard_20152023v1.0.1.csv')

# Calculate average price for each PLU code
plu_codes = ['PLU 4046', 'PLU 4225', 'PLU 4770']
plu_columns = ['plu4046', 'plu4225', 'plu4770']
avg_prices = {plu: avocado_data[column].mean() for plu, column in zip(plu_codes, plu_columns)}

# Create a bar plot for average price
plt.figure(figsize=(6, 6))
barplot = sns.barplot(x=list(avg_prices.keys()), y=list(avg_prices.values()), hue=list(avg_prices.keys()), palette='viridis', legend=False)
plt.title('Average Price for PLU Codes')
plt.xlabel('PLU Codes')
plt.ylabel('Average Price')

# Add data values on top of the bars
for i, (plu, price) in enumerate(avg_prices.items()):barplot.text(i, price + 0.1, f'{price:.2f}', ha='center', va='bottom')
plt.show()
```
